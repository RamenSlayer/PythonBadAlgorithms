```python
def SqRt(n, itr=50):
    step = 0
    left = 0
    right = n
    mid = 0
    while step < itr:
        mid = (left + right)/2
        tmp = mid*mid
        if tmp == n:
            return mid
        elif tmp > n:
            right = mid
        elif tmp < n:
            left = mid
        step += 1
        print(f"left = {left}\nright = {right}\n")
    return mid
```

```python
SqRt(2, 50)
```
