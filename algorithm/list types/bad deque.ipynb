```python
class Deque():
    def __init__(self):
        self.deque = []
        self.Size = 0
    def push_front(self, n):
        self.deque.append(n)
        self.Size += 1
    def push_back(self, n):
        self.deque.insert(0, n)
        self.Size += 1
    def pop_front(self):
        if self.Size > 0:
            self.Size -= 1
            return self.deque.pop()
        else:
            return "error"
    def pop_back(self):
        if self.Size > 0:
            self.Size -= 1
            return self.deque.pop(0)
        else:
            return "error"
    def front(self):
        if self.Size > 0:
            return self.deque[-1]
        else:
            return "error"
    def back(self):
        if self.Size > 0:
            return self.deque[0]
        else:
            return "error"
    def size(self):
        return self.Size
    def clear(self):
        self.deque = []
        self.Size = 0
```

```python
d = Deque()
while True:
    inp = input().split()
    cmd = inp[0]
    if cmd == "push_front":
        d.push_front(inp[1])
        print("ok")
    elif cmd == "push_back":
        d.push_back(inp[1])
        print("ok")
    elif cmd == "pop_front":
        print(d.pop_front())
    elif cmd == "pop_back":
        print(d.pop_back())
    elif cmd == "front":
        print(d.front())
    elif cmd == "back":
        print(d.back())
    elif cmd == "size":
        print(d.size())
    elif cmd == "clear":
        d.clear()
        print("ok")
    elif cmd == "exit":
        print("bye")
        break
```
